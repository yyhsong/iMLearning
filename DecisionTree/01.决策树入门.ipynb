## 决策树 Decision Tree

- 决策树既可以用于分类，又可以用于回归。
- 决策树是一个树结构（二叉树或非二叉树），每个非叶节点对应一个特征，该节点的每个分支代表这个特征的一个取值，而每个叶节点存放一个类别或一个回归函数。
- 使用决策树进行决策的过程，就是从根节点开始，提取出待分类项中相应的特征，按照其值选择输出分支，依次向下，直到到达叶子节点，将叶子节点存放的类别或者回归函数的运算结果作为输出（决策）结果。
- 非叶子节点（含根节点）统称为决策节点，每个节点对应一个条件判断，这个判断的条件叫做特征。
- 决策节点的重要性是不一样的，最重要的根节点，越靠近根节点也越重要。

## 构建决策树

获得模型的过程叫做训练，可以通过以下几步得到一棵决策树：
1. 准备若干的训练数据（假设有 m 个样本）；
2. 标明每个样本预期的类别；
3. 人为选取一些特征（即决策条件）；
4. 为每个训练样本对应所有需要的特征生成相应值——数值化特征；
5. 将通过上面的1-4步获得的训练数据输入给训练算法，训练算法通过一定的原则，决定各个特征的重要性程度，然后按照决策重要性从高到底，生成决策树。

决策树的构建过程是一个迭代的过程，每次迭代中，采用不同特征作为分裂点，来将样本数据划分成不同的类别，被用作分裂点的特征叫做分裂特征。

## 决策树算法

- ID3算法（Iterative Dichotomiser 3）
    - 该算法的核心是以信息增益为度量，选择分裂后信息增益最大的特征进行分裂。
    - 信息熵，熵表示信息的混乱程度，信息越混乱，熵值越大。
    - 缺点：ID3一般会优先选择取值种类较多的特征作为分裂特征。
- C4.5算法
    - 用信息增益率（Gain Ratio），用比例而不是单纯的量，作为选择分支的标准。
    - 信息增益率通过引入一个被称为分裂信息（Split Information）的项，来惩罚取值可能性较多的特征。
- CART算法 （Classification and Regression Tree）
    - 特征选取依据不是信息增益量或者增益率，而是Gini系数（Gini Coefficient），每次选择Gini系数最小的特征作为最优切分点。
    - Gini系数是一个比例数，取值在0到1之间，Gini指数（Gini Index）是Gini系数乘100倍作百分比表示，两者本质上是一致的。
    - CART是一棵严格的二叉树，每次分裂只做二分。
    - 回归树和分类树的区别在于最终的输出值到底是连续的还是离散的，而特征（即决策点分裂条件）无论是连续还是离散，都要被当做离散值来处理，而且都是被转化为二分类特征来进行处理。
